```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from torchvision import datasets, transforms
from PIL import Image
import numpy as np

data_dir = r"C:\Users\avish\Documents\animal_classifier_unified_mentor\data\processed\train"
class_names = sorted(os.listdir(data_dir))
print("Classes:", class_names)
```

```python
class_counts = {cls: len(os.listdir(os.path.join(data_dir, cls))) for cls in class_names}
sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()))
plt.xticks(rotation=45)
plt.title("Class Distribution")
plt.ylabel("Number of Images")
plt.xlabel("Animal Classes")
plt.tight_layout()
plt.show()
```

```python
ncols = 5
nrows = int(np.ceil(len(class_names) / ncols))
plt.figure(figsize=(15, 3 * nrows))

for idx, cls in enumerate(class_names):
    img_folder = os.path.join(data_dir, cls)
    img_file = os.listdir(img_folder)[0]
    img_path = os.path.join(img_folder, img_file)
    img = Image.open(img_path)
    
    plt.subplot(nrows, ncols, idx + 1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
